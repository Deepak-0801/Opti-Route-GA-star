# Brute Force Method

---

This Python script provides a basic implementation to solve the Traveling Salesman Problem (TSP) using a brute-force approach. The TSP involves finding the shortest possible route that visits a given set of cities and returns to the starting city. The script calculates all possible routes and identifies the one with the minimum total distance.

## Code Overview

- The `calculate_total_distance` function computes the total distance of a given tour based on a provided distance matrix.
- The `brute_force_tsp` function explores all permutations of cities to find the optimal tour with the minimum total distance.
- Example usage of the script is demonstrated with a sample distance matrix representing distances between different cities.

```python
from itertools import permutations

def calculate_total_distance(tour, distances):
    total_distance = 0
    num_cities = len(tour)

    for i in range(num_cities - 1):
        total_distance += distances[tour[i]][tour[i + 1]]

    # Return to the starting city
    total_distance += distances[tour[-1]][tour[0]]

    return total_distance

def brute_force_tsp(distances):
    num_cities = len(distances)
    cities = list(range(num_cities))
    all_routes = []
    min_distance = float('inf')
    best_tour = None

    for perm in permutations(cities):
        current_distance = calculate_total_distance(perm, distances)
        all_routes.append((perm, current_distance))

        if current_distance < min_distance:
            min_distance = current_distance
            best_tour = perm

    return all_routes, best_tour, min_distance

# Example usage:
distances = [
    [0, 29, 20, 21, 16, 31, 100, 12, 4, 31],
    [29, 0, 15, 29, 28, 40, 72, 21, 29, 41],
    [20, 15, 0, 15, 14, 25, 81, 9, 23, 27],
    [21, 29, 15, 0, 4, 12, 92, 12, 25, 13],
    [16, 28, 14, 4, 0, 16, 94, 9, 20, 16],
    [31, 40, 25, 12, 16, 0, 95, 24, 36, 3],
    [100, 72, 81, 92, 94, 95, 0, 90, 101, 99],
    [12, 21, 9, 12, 9, 24, 90, 0, 17, 42],
    [4, 29, 23, 25, 20, 36, 101, 17, 0, 35],
    [31, 41, 27, 13, 16, 3, 99, 42, 35, 0]
]

all_routes, best_tour, min_distance = brute_force_tsp(distances)

print("All Routes:")
for route, distance in all_routes:
    print(f"Route: {route}, Distance: {distance}")

print("\nBest Route:", best_tour)
print("Minimum Distance:", min_distance)
```

This extended Python script now utilizes a Genetic Algorithm (GA) to solve the Traveling Salesman Problem (TSP). The GA incorporates methods such as individual creation, crossover (Order Crossover - OX), mutation (Swap Mutation), and selection to evolve a population of routes over several generations. The goal is to find the optimal route with the minimum total distance.

## Code Overview

- The `create_individual` function generates a random tour as an individual.
- The `crossover` function employs the Order Crossover (OX) method to create offspring from two parent individuals.
- The `mutate` function applies Swap Mutation to introduce variation in an individual.
- The `genetic_algorithm` function orchestrates the GA process, evolving a population over a specified number of generations.


```python
import random
import math
import heapq
import matplotlib.pyplot as plt

def calculate_total_distance(tour, distances):
    total_distance = 0
    num_cities = len(tour)

    for i in range(num_cities - 1):
        total_distance += distances[tour[i]][tour[i + 1]]

    # Return to the starting city
    total_distance += distances[tour[-1]][tour[0]]

    return total_distance

def create_individual(num_cities):
    individual = list(range(num_cities))
    random.shuffle(individual)
    return individual

def crossover(parent1, parent2):
    # Order Crossover (OX) method
    start = random.randint(0, len(parent1) - 1)
    end = random.randint(start + 1, len(parent1))
    child = [-1] * len(parent1)

    # Copy a segment from parent1
    child[start:end] = parent1[start:end]

    # Fill in the remaining positions with genes from parent2
    index = end
    for gene in parent2 + parent2:
        if gene not in child:
            child[index % len(parent1)] = gene
            index += 1

        if -1 not in child:
            break

    return child

def mutate(individual):
    # Swap mutation
    idx1, idx2 = random.sample(range(len(individual)), 2)
    individual[idx1], individual[idx2] = individual[idx2], individual[idx1]
    return individual

def a_star_swap_optimization(initial_tour, distances):
    num_cities = len(initial_tour)

    def heuristic(tour):
        return calculate_total_distance(tour, distances)

    def neighbors(tour):
        tour_neighbors = []
        for i in range(num_cities):
            for j in range(i + 1, num_cities):
                neighbor = tour[:]
                neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
                tour_neighbors.append(neighbor)
        return tour_neighbors

    priority_queue = [(heuristic(initial_tour), initial_tour)]

    while priority_queue:
        _, current_tour = heapq.heappop(priority_queue)

        for neighbor in neighbors(current_tour):
            neighbor_cost = heuristic(neighbor)
            heapq.heappush(priority_queue, (neighbor_cost, neighbor))

        if heuristic(current_tour) < heuristic(initial_tour):
            initial_tour = current_tour

    return initial_tour

def genetic_algorithm_with_a_star(distances, population_size=50, generations=1000, a_star_percentage=0):
    num_cities = len(distances)
    population = [create_individual(num_cities) for _ in range(population_size)]
    best_distances = []

    for generation in range(generations):
        population = sorted(population, key=lambda x: calculate_total_distance(x, distances))

        # Select the top 50% as parents
        parents = population[:population_size // 2]

        # Create offspring through crossover and mutation
        offspring = []
        while len(offspring) < population_size - len(parents):
            parent1, parent2 = random.sample(parents, 2)
            child = crossover(parent1, parent2)
            child = mutate(child)
            offspring.append(child)

        population = parents + offspring

        # Apply A* swap optimization to a subset of the population
        a_star_population_size = int(a_star_percentage * population_size)
        a_star_candidates = random.sample(population, a_star_population_size)

        for candidate in a_star_candidates:
            optimized_candidate = a_star_swap_optimization(candidate, distances)
            population[population.index(candidate)] = optimized_candidate

        # Record the best distance in this generation
        best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
        best_distance = calculate_total_distance(best_route, distances)
        best_distances.append(best_distance)

    # Return the best route found
    best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
    best_distance = calculate_total_distance(best_route, distances)

    return best_route, best_distance, best_distances

# Example usage:
distances = [
    [0, 29, 20, 21, 16, 31, 100, 12, 4, 31],
    [29, 0, 15, 29, 28, 40, 72, 21, 29, 41],
    [20, 15, 0, 15, 14, 25, 81, 9, 23, 27],
    [21, 29, 15, 0, 4, 12, 92, 12, 25, 13],
    [16, 28, 14, 4, 0, 16, 94, 9, 20, 16],
    [31, 40, 25, 12, 16, 0, 95, 24, 36, 3],
    [100, 72, 81, 92, 94, 95, 0, 90, 101, 99],
    [12, 21, 9, 12, 9, 24, 90, 0, 17, 42],
    [4, 29, 23, 25, 20, 36, 101, 17, 0, 35],
    [31, 41, 27, 13, 16, 3, 99, 42, 35, 0]
]

best_route, best_distance, best_distances_over_generations = genetic_algorithm_with_a_star(distances)

# Plot the performance over generations
plt.plot(best_distances_over_generations, label='Best Distance')
plt.xlabel('Generation')
plt.ylabel('Distance')
plt.title('Genetic Algorithm')
plt.legend()
plt.show()

print("Best Route:", best_route)
print("Minimum Distance:", best_distance)
```

This extended Python script combines a Genetic Algorithm (GA) with A* Swap Optimization to solve the Traveling Salesman Problem (TSP). - The `genetic_algorithm_with_a_star` function integrates GA with A* optimization, allowing for more effective exploration.

```python
import random
import math
import heapq
import matplotlib.pyplot as plt

def calculate_total_distance(tour, distances):
    total_distance = 0
    num_cities = len(tour)

    for i in range(num_cities - 1):
        total_distance += distances[tour[i]][tour[i + 1]]

    # Return to the starting city
    total_distance += distances[tour[-1]][tour[0]]

    return total_distance

def create_individual(num_cities):
    individual = list(range(num_cities))
    random.shuffle(individual)
    return individual

def crossover(parent1, parent2):
    # Order Crossover (OX) method
    start = random.randint(0, len(parent1) - 1)
    end = random.randint(start + 1, len(parent1))
    child = [-1] * len(parent1)

    # Copy a segment from parent1
    child[start:end] = parent1[start:end]

    # Fill in the remaining positions with genes from parent2
    index = end
    for gene in parent2 + parent2:
        if gene not in child:
            child[index % len(parent1)] = gene
            index += 1

        if -1 not in child:
            break

    return child

def mutate(individual, mutation_rate):
    # Swap mutation
    for _ in range(int(mutation_rate * len(individual))):
        idx1, idx2 = random.sample(range(len(individual)), 2)
        individual[idx1], individual[idx2] = individual[idx2], individual[idx1]
    return individual

def a_star_swap_optimization(initial_tour, distances):
    num_cities = len(initial_tour)

    def heuristic(tour):
        return calculate_total_distance(tour, distances)

    def neighbors(tour):
        tour_neighbors = []
        for i in range(num_cities):
            for j in range(i + 1, num_cities):
                neighbor = tour[:]
                neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
                tour_neighbors.append(neighbor)
        return tour_neighbors

    current_tour = initial_tour

    while True:
        improved = False
        for neighbor in neighbors(current_tour):
            if heuristic(neighbor) < heuristic(current_tour):
                current_tour = neighbor
                improved = True

        if not improved:
            break

    return current_tour

def genetic_algorithm_with_a_star(distances, population_size=50, generations=1000, a_star_percentage=0.1, mutation_rate=0.1, plateau_limit=500):
    num_cities = len(distances)
    population = [create_individual(num_cities) for _ in range(population_size)]
    best_distances = []

    plateau_count = 0
    previous_best_distance = float('inf')

    for generation in range(generations):
        population = sorted(population, key=lambda x: calculate_total_distance(x, distances))

        # Select the top 50% as parents
        parents = population[:population_size // 2]

        # Create offspring through crossover and mutation
        offspring = []
        while len(offspring) < population_size - len(parents):
            parent1, parent2 = random.sample(parents, 2)
            child = crossover(parent1, parent2)
            child = mutate(child, mutation_rate)
            offspring.append(child)

        population = parents + offspring

        # Apply A* swap optimization to a subset of the population
        a_star_population_size = int(a_star_percentage * population_size)
        a_star_candidates = random.sample(population, a_star_population_size)

        for i in range(a_star_population_size):
            a_star_candidates[i] = a_star_swap_optimization(a_star_candidates[i], distances)

        population[:a_star_population_size] = a_star_candidates

        # Record the best distance in this generation
        best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
        best_distance = calculate_total_distance(best_route, distances)
        best_distances.append(best_distance)

        # Check for plateau
        if best_distance == previous_best_distance:
            plateau_count += 1
        else:
            plateau_count = 0

        if plateau_count == plateau_limit:
            print(f"Terminating due to plateauing for {plateau_limit} generations.")
            break

        previous_best_distance = best_distance

    # Return the best route found
    best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
    best_distance = calculate_total_distance(best_route, distances)

    return best_route, best_distance, best_distances

distances = [
    [0, 29, 20, 21, 16, 31, 100, 12, 4, 31],
    [29, 0, 15, 29, 28, 40, 72, 21, 29, 41],
    [20, 15, 0, 15, 14, 25, 81, 9, 23, 27],
    [21, 29, 15, 0, 4, 12, 92, 12, 25, 13],
    [16, 28, 14, 4, 0, 16, 94, 9, 20, 16],
    [31, 40, 25, 12, 16, 0, 95, 24, 36, 3],
    [100, 72, 81, 92, 94, 95, 0, 90, 101, 99],
    [12, 21, 9, 12, 9, 24, 90, 0, 17, 42],
    [4, 29, 23, 25, 20, 36, 101, 17, 0, 35],
    [31, 41, 27, 13, 16, 3, 99, 42, 35, 0]
]

population_size = 100
mutation_rate = 0.6

generations = 1000
plateau_limit = 5000

best_route, best_distance, best_distances_over_generations = genetic_algorithm_with_a_star(
    distances,
    population_size=population_size,
    generations=generations,
    mutation_rate=mutation_rate,
    plateau_limit=plateau_limit
)

# Plot the performance over generations
plt.plot(range(len(best_distances_over_generations)), best_distances_over_generations, label='Best Distance')
plt.xlabel('Generation')
plt.ylabel('Distance')
plt.title('Genetic Algorithm with A* Swap Optimization')
plt.legend()
plt.show()

print("Best Route:", best_route)
print("Minimum Distance:", best_distance)
```

# A* Mutation (Low Granularity)

This enhanced Python script integrates Genetic Algorithm (GA) with A* Swap Optimization and introduces a plateau detection mechanism to solve the Traveling Salesman Problem (TSP). The GA evolves a population of routes over several generations, applying A* optimization to a subset of individuals and detecting plateaus to optimize convergence.

## Code Overview

- The `local_search_swap` function performs a local search on a tour using swap operations.
- The `mutate_with_a_star` function applies A*-like local search to optimize individuals during mutation.
- The `genetic_algorithm_with_a_star` function incorporates A* optimization, plateau detection, and visualization of performance.



```python
import random
import matplotlib.pyplot as plt

def calculate_total_distance(tour, distances):
    total_distance = 0
    num_cities = len(tour)

    for i in range(num_cities - 1):
        total_distance += distances[tour[i]][tour[i + 1]]

    # Return to the starting city
    total_distance += distances[tour[-1]][tour[0]]

    return total_distance

def create_individual(num_cities):
    individual = list(range(num_cities))
    random.shuffle(individual)
    return individual

def crossover(parent1, parent2):
    # Order Crossover (OX) method
    start = random.randint(0, len(parent1) - 1)
    end = random.randint(start + 1, len(parent1))
    child = [-1] * len(parent1)

    # Copy a segment from parent1
    child[start:end] = parent1[start:end]

    # Fill in the remaining positions with genes from parent2
    index = end
    for gene in parent2 + parent2:
        if gene not in child:
            child[index % len(parent1)] = gene
            index += 1

        if -1 not in child:
            break

    return child

def local_search_swap(tour, distances):
    num_cities = len(tour)
    improved = True

    while improved:
        improved = False

        for i in range(num_cities - 1):
            for j in range(i + 1, num_cities):
                new_tour = tour[:]
                new_tour[i], new_tour[j] = new_tour[j], new_tour[i]
                new_distance = calculate_total_distance(new_tour, distances)

                if new_distance < calculate_total_distance(tour, distances):
                    tour = new_tour
                    improved = True

    return tour

def mutate_with_a_star(individual, distances, mutation_rate):
    if random.random() < mutation_rate:
        # Apply A*-like local search to optimize the individual
        optimized_individual = local_search_swap(individual, distances)
        return optimized_individual
    else:
        return individual

def genetic_algorithm_with_a_star(distances, population_size=50, generations=1000, mutation_rate=0.1, plateau_limit=500):
    num_cities = len(distances)
    population = [create_individual(num_cities) for _ in range(population_size)]
    best_distances = []

    plateau_count = 0
    previous_best_distance = float('inf')

    for generation in range(generations):
        population = sorted(population, key=lambda x: calculate_total_distance(x, distances))

        # Select the top 50% as parents
        parents = population[:population_size // 2]

        # Create offspring through crossover and mutation
        offspring = []
        for parent in parents:
            child = mutate_with_a_star(parent, distances, mutation_rate)
            offspring.append(child)

        population = parents + offspring

        # Record the best distance in this generation
        best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
        best_distance = calculate_total_distance(best_route, distances)
        best_distances.append(best_distance)

        # Check for plateau
        if best_distance == previous_best_distance:
            plateau_count += 1
        else:
            plateau_count = 0

        if plateau_count == plateau_limit:
            print(f"Terminating due to plateauing for {plateau_limit} generations.")
            break

        previous_best_distance = best_distance

    # Return the best route found
    best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
    best_distance = calculate_total_distance(best_route, distances)

    return best_route, best_distance, best_distances

# Example usage:
distances = [
    [0, 29, 20, 21, 16, 31, 100, 12, 4, 31],
    [29, 0, 15, 29, 28, 40, 72, 21, 29, 41],
    [20, 15, 0, 15, 14, 25, 81, 9, 23, 27],
    [21, 29, 15, 0, 4, 12, 92, 12, 25, 13],
    [16, 28, 14, 4, 0, 16, 94, 9, 20, 16],
    [31, 40, 25, 12, 16, 0, 95, 24, 36, 3],
    [100, 72, 81, 92, 94, 95, 0, 90, 101, 99],
    [12, 21, 9, 12, 9, 24, 90, 0, 17, 42],
    [4, 29, 23, 25, 20, 36, 101, 17, 0, 35],
    [31, 41, 27, 13, 16, 3, 99, 42, 35, 0]
]

population_size = 100
mutation_rate = 0.02

generations = 1000
plateau_limit = 10000

best_route, best_distance, best_distances_over_generations = genetic_algorithm_with_a_star(
    distances,
    population_size=population_size,
    generations=generations,
    mutation_rate=mutation_rate,
    plateau_limit=plateau_limit
)

# Plot the performance over generations
plt.plot(range(len(best_distances_over_generations)), best_distances_over_generations, label='Best Distance')
plt.xlabel('Generation')
plt.ylabel('Distance')
plt.title('Genetic Algorithm with A* Single City Mutation')
plt.legend()
plt.show()

print("Best Route:", best_route)
print("Minimum Distance:", best_distance)
```

# A* Mutation (High Granularity)

This Python script extends the Traveling Salesman Problem (TSP) solver by combining a Genetic Algorithm (GA) with A* Multi-City Swap Optimization. The algorithm evolves a population of routes over several generations, applying A* optimization to improve individuals. ## Code Overview

- The `a_star_multi_city_swap_optimization` function applies A* search to multi-city swap optimization for a given individual.
- The `mutate` function integrates A* multi-city swap optimization into the mutation step of the GA.
- The `genetic_algorithm_with_parameters` function allows users to customize key parameters and visualizes the performance over generations.



```python
import random
import matplotlib.pyplot as plt

def calculate_total_distance(tour, distances):
    total_distance = 0
    num_cities = len(tour)

    for i in range(num_cities - 1):
        total_distance += distances[tour[i]][tour[i + 1]]

    # Return to the starting city
    total_distance += distances[tour[-1]][tour[0]]

    return total_distance

def create_individual(num_cities):
    individual = list(range(num_cities))
    random.shuffle(individual)
    return individual

def crossover(parent1, parent2):
    # Order Crossover (OX) method
    start = random.randint(0, len(parent1) - 1)
    end = random.randint(start + 1, len(parent1))
    child = [-1] * len(parent1)

    # Copy a segment from parent1
    child[start:end] = parent1[start:end]

    # Fill in the remaining positions with genes from parent2
    index = end
    for gene in parent2 + parent2:
        if gene not in child:
            child[index % len(parent1)] = gene
            index += 1

        if -1 not in child:
            break

    return child

def a_star_multi_city_swap_optimization(initial_tour, distances):
    num_cities = len(initial_tour)

    def heuristic(tour):
        return calculate_total_distance(tour, distances)

    def neighbors(tour):
        tour_neighbors = []
        for i in range(num_cities):
            for j in range(i + 1, num_cities):
                neighbor = tour[:]
                neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
                tour_neighbors.append(neighbor)
        return tour_neighbors

    current_tour = initial_tour

    while True:
        improved = False
        for neighbor in neighbors(current_tour):
            if heuristic(neighbor) < heuristic(current_tour):
                current_tour = neighbor
                improved = True

        if not improved:
            break

    return current_tour

def mutate(individual):
    # A* multi-city swap optimization
    individual = a_star_multi_city_swap_optimization(individual, distances)
    return individual

def genetic_algorithm_with_parameters(distances, population_size=50, generations=1000, mutation_rate=0.1, plateau_limit=500):
    num_cities = len(distances)
    population = [create_individual(num_cities) for _ in range(population_size)]
    best_distances = []

    plateau_count = 0
    previous_best_distance = float('inf')

    for generation in range(generations):
        population = sorted(population, key=lambda x: calculate_total_distance(x, distances))

        # Select the top 50% as parents
        parents = population[:population_size // 2]

        # Create offspring through crossover and mutation
        offspring = []
        while len(offspring) < population_size - len(parents):
            parent1, parent2 = random.sample(parents, 2)
            child = crossover(parent1, parent2)
            child = mutate(child)
            offspring.append(child)

        population = parents + offspring

        # Record the best distance in this generation
        best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
        best_distance = calculate_total_distance(best_route, distances)
        best_distances.append(best_distance)

        # Check for plateau
        if best_distance == previous_best_distance:
            plateau_count += 1
        else:
            plateau_count = 0

        if plateau_count == plateau_limit:
            print(f"Terminating due to plateauing for {plateau_limit} generations.")
            break

        previous_best_distance = best_distance

    # Plot the performance over generations
    plt.plot(range(len(best_distances)), best_distances, label='Best Distance')
    plt.xlabel('Generation')
    plt.ylabel('Distance')
    plt.title('Genetic Algorithm with A* Multi City Mutation')
    plt.legend()
    plt.show()

    # Return the best route found
    best_route = min(population, key=lambda x: calculate_total_distance(x, distances))
    best_distance = calculate_total_distance(best_route, distances)

    return best_route, best_distance

# Example usage with custom parameters:
distances = [
    [0, 29, 20, 21, 16, 31, 100, 12, 4, 31],
    [29, 0, 15, 29, 28, 40, 72, 21, 29, 41],
    [20, 15, 0, 15, 14, 25, 81, 9, 23, 27],
    [21, 29, 15, 0, 4, 12, 92, 12, 25, 13],
    [16, 28, 14, 4, 0, 16, 94, 9, 20, 16],
    [31, 40, 25, 12, 16, 0, 95, 24, 36, 3],
    [100, 72, 81, 92, 94, 95, 0, 90, 101, 99],
    [12, 21, 9, 12, 9, 24, 90, 0, 17, 42],
    [4, 29, 23, 25, 20, 36, 101, 17, 0, 35],
    [31, 41, 27, 13, 16, 3, 99, 42, 35, 0]
]

custom_population_size = 100
custom_generations = 1000
custom_mutation_rate = 0.4
custom_plateau_limit = 1000

best_route, best_distance = genetic_algorithm_with_parameters(
    distances,
    population_size=custom_population_size,
    generations=custom_generations,
    mutation_rate=custom_mutation_rate,
    plateau_limit=custom_plateau_limit
)

print("Best Route:", best_route)
print("Minimum Distance:", best_distance)
```
